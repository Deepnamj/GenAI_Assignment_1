1.Using the products list and price_dict, create a list of tuples named catalog where each tuple is (product_name, price, category).

2.From catalog, create a new dictionary category_to_products that maps each category to a list of product names in that category.

3.Print all products that belong to the category that has the maximum number of products.


```python
# 1. Create catalog dynamically: list of tuples (product_name, price, category)
catalog = [
    (p, price_dict[p], categories[i])  # create tuple: (product_name, price, category)
    for i, p in enumerate(products)    # enumerate products to get both index and name
    if p in price_dict                 # only include product if it exists in price_dict
]
print("Catalog:", catalog)  # display the full catalog


# 2. Create category_to_products dictionary dynamically
category_to_products = {}  # initialize empty dictionary
for product, price, category in catalog:  # loop through each item in catalog
    if category not in category_to_products:  # check if category key exists
        category_to_products[category] = []  # if not, initialize with empty list
    category_to_products[category].append(product)  # append product to its category list

print("Category to products mapping:", category_to_products)  # show mapping


# 3. Find and print products in category with max number of products
max_count = max(len(p_list) for p_list in category_to_products.values())  # find the largest category size
for category, p_list in category_to_products.items():  # loop through categories
    if len(p_list) == max_count:  # check if this category has the maximum number of products
        print("Products in category with max items:", p_list)  # print the products in that category
```
