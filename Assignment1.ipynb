```python
price_dict = {"apple":10, "banana":15, "orange":20, "grape":25, "kiwi":30, "melon":35}

price_dict["pear"] = 40
print(price_dict)
price_dict["apple"] = 12
print(price_dict)
del price_dict["banana"]
print(price_dict)

avg_price = sum(price_dict.values()) / len(price_dict)
print(avg_price)

max_price = max(price_dict.values())
print(max_price)    
min_price = min(price_dict.values())
print(min_price)

```
